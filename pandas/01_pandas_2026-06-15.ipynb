# Pandas Series
Source: W3 schools
It's a column in a table (one dimensional array holding data of *any* type)

```python
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
# If labels are not specified, they are labelled with their index numbers
# First value has index 9

print(myvar[0])

# Making your own labels:

myvar = pd.Series(a, index = ['x', 'y', 'z'])

print(myvar)
```

```python
# Key/value objects can be used like a dictionary when creating a series????
# "create a simple pandas series from a dictionary:"

# ok wait so key refers to label value refers to data?
# so if i have a dictionary with 2 dimensions, one dimension is the label which becomes key and the other is the value?


calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)

myvar2 = pd.Series(calories, index = ["day1", "day2"])
print(myvar2)

# Dataframes
data_frame = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data_frame)
print(myvar)
```

# Dataframes

```python
data_frame = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

# load data into DataFrame object:

df = pd.DataFrame(data_frame)

print(df)
```

```python
# What is a dataframe? What is the data type of data before it becomes a dataframe?

print(type(data_frame))
print(type(df))
```
